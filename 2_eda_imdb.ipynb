### 24. Crea la matriz de correlación del dataframe de películas

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python
import matplotlib.pyplot as plt

plt.figure()
plt.matshow(matriz_correlacion)
plt.colorbar()
plt.xticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns, rotation=45)
plt.yticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns)
plt.title('Matriz de Correlación de Películas', pad=20)
plt.show()
```
